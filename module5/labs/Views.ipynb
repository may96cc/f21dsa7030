## Clarity through Column Renames 
One of the issues we see is that the `id` column exists twice.
First from the `visited.id` column, then from the `person.id` column.

We need to fix this to have unambiguous columns.

We also see that the `person` column of the output is redundant with the `person.id` column. 

Let's update our SQL to make the results clearer.

````python
SQL = """SELECT v.id as visit_id, 
    s.quant, s.reading,
    v.site, v.dated as visit_date,
    p.id as person_id, p.personal as firstname, p.family as lastname
FROM {pawprint}.survey s 
JOIN {pawprint}.visited v ON (v.id::int = s.taken::int) 
JOIN {pawprint}.person p ON (s.person=p.id);
"""
# Slide the SSO from variable in first cell into the SQL string.
SQL = SQL.format(pawprint=SSO)
print(SQL)
````

````python
# Use the SQL Alchemy to get a connection and read SQL into a table
with engine.connect() as connection:
    df = pd.read_sql_query(SQL, connection)

#  Show to the first 20 rows of the 19 rows
df.head(20)
````

## Now that we have this _report_ from the database, let's make this accessible for future use without re-writing the SQL.

To do this, we will create a view, which is basically a stored, named `table expression`.

Recalling a select statement is the definition of a table, a table expression, then:  
`CREATE OR REPLACE VIEW SSO.NAME AS SELECT_STATEMENT`

````python
SQL = """CREATE OR REPLACE VIEW 
{pawprint}.survey_view
AS
SELECT v.id as visit_id, 
    s.quant, s.reading,
    v.site, v.dated as visit_date,
    p.id as person_id, p.personal as firstname, p.family as lastname
FROM {pawprint}.survey s 
JOIN {pawprint}.visited v ON (v.id::int = s.taken::int) 
JOIN {pawprint}.person p ON (s.person=p.id);
"""
# Slide the SSO from variable in first cell into the SQL string.
SQL = SQL.format(pawprint=SSO)
print(SQL)
````

````python
result = engine.execute(SQL)
print(result)
````

## Selecting data from a view
#### ... is the same as selecting data from single table.

````python
with engine.connect() as connection:
    df = pd.read_sql_query("SELECT * FROM {pawprint}.survey_view".format(pawprint=SSO), connection)

#  Show to the first 20 rows of the 19 rows
df.head(20)
````
